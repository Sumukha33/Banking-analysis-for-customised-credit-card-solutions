```python
merged_df=pd.merge(df1,df2, on='customer_id',how='inner')
merged_df
```

```python
spend_per_customer=merged_df.groupby('customer_id')['spend'].sum()
spend_per_customer
```

```python
spend_type=df2.groupby('payment_type')['spend'].sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 6))
plt.pie(spend_type, labels=spend_type.index, autopct='%1.1f%%', startangle=140)
plt.title('spendings by payment type')
plt.show()
```

```python
total_income_by_gender = df1.groupby('gender')['avg_income'].sum()
```

```python
plt.pie(total_income_by_gender, labels=total_income_by_gender.index, autopct='%1.1f%%', startangle=140)
plt.title('Income by Gender')
plt.show()
```

```python
total_spend=df2.groupby('category')['spend'].sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 8))
plt.pie(total_spend, labels=total_spend.index, autopct='%1.1f%%', startangle=140)
plt.title('Spendings by category')
plt.show()
```

```python
city_spend=merged_df.groupby('city')['spend'].sum()
```

```python
plt.pie(city_spend, labels=city_spend.index, autopct='%1.1f%%', startangle=140)
plt.title('Spendings in city')
plt.show()
```

```python
income_city = df1.groupby('city')['avg_income'].sum()
```

```python
occupation_spend = merged_df.groupby('occupation')['spend'].sum().reset_index()
```

```python
plt.bar(occupation_spend['occupation'], occupation_spend['spend'])
plt.xlabel('Occupation')
plt.ylabel('Total Spend')
plt.title('Total Spending by Occupation')
plt.xticks(rotation=45)  
plt.tight_layout()  
plt.show()
```

```python
occupation_spend = merged_df.groupby('age_group')['spend'].sum().reset_index()
```

```python
plt.bar(occupation_spend['age_group'], occupation_spend['spend'])
plt.xlabel('age group')
plt.ylabel('Total Spend')
plt.title('Total Spending by age ')
plt.xticks(rotation=45)  
plt.tight_layout()  
plt.show()
```
